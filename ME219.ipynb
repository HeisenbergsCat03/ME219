```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(15, 9), dpi=80)


```

```python
x=np.array([1,5,10,15])
y = np.array([46.0,56.0,102.0,326.0])

for i in (0,3,1,2):
    y[i]=(2*np.pi*60)/y[i]
    print(y)
plt.xlim(0,16)
plt.ylim(0,10)
plt.plot(x,y)
plt.scatter(x,y,color ='red')
plt.xlabel("Time passed in seconds")
plt.ylabel("Angular frequency in radians per sec")
plt.title("Decay of angular velocity with time at 120 bpm")

```
